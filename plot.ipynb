```python
# this line prepares IPython for working with matplotlib
%matplotlib inline  

import numpy as np
import matplotlib.pyplot as plt  

# x = [1, 3, 5, 10] 
x = [1, 2, 3, 4]
acc = [0.654266, 0.654266, 0.754953, 0.784877]
fp = [0.301349, 0.301349, 0.113943, 0.056972]
fn = [0.533898, 0.533898, 0.800847, 0.885593]
plt.plot(x, acc, label = 'Accuracy')
plt.plot(x, fp, label = 'False Positive') 
plt.plot(x, fn, label = 'False Negative')
# your code here (plus you may need to modify the previous two lines)
# label for axis
plt.xlabel('rounds')
plt.ylabel('rate')

# legend
# plt.legend(loc='lower right')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
fig = plt.gcf()
plt.savefig('result.png')
```
